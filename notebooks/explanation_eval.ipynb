```python
# Function to adjust bar widths
def adjust_barwidth(ax, new_width):
    for patch in ax.patches:
        current_width = patch.get_width()
        diff = current_width - new_width

        # Change the bar width
        patch.set_width(new_width)

        # Recenter the bar
        patch.set_x(patch.get_x() + diff * .5)
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame with an additional column
data = {
    'numbers': [1, 2, 3, 4, 5, 1, 2, 2, 3, 4, 5, 5, 5, 3, 2, 1, 4, 5, 3, 2],
    'additional': [1, 3, 3, 5, 4, 2, 4, 1, 5, 4, 3, 4, 4, 2, 5, 3, 2, 1, 4, 5]  # New data for the additional column
}
df = pd.DataFrame(data)

# Set up the matplotlib figure
fig = plt.figure(figsize=(14, 6))  # Adjust the figure size to accommodate two plots

# Function to adjust bar widths
def adjust_barwidth(ax, new_width):
    for patch in ax.patches:
        current_width = patch.get_width()
        diff = current_width - new_width

        # Change the bar width
        patch.set_width(new_width)

        # Recenter the bar
        patch.set_x(patch.get_x() + diff * .5)

# First subplot for 'numbers' column
ax1 = plt.subplot(1, 2, 1)  # 1 row, 2 columns, 1st subplot
sns.countplot(x='numbers', data=df, palette='viridis')
plt.title('Count of Numbers in Column')
plt.ylabel('Count')
plt.xlabel('Number')

# Adjust bar width
adjust_barwidth(ax1, new_width=0.7)

# Second subplot for 'additional' column
ax2 = plt.subplot(1, 2, 2)  # 1 row, 2 columns, 2nd subplot
sns.countplot(x='additional', data=df, palette='viridis')
plt.title('Count of Additional Column')
plt.ylabel('Count')
plt.xlabel('Additional Number')

# Adjust bar width
adjust_barwidth(ax2, new_width=0.97)

ax2.set(yticklabels=[])  
ax2.set(ylabel=None)  # remove the y-axis label
ax2.tick_params(left=False)  # remove the ticks

# Adjust layout to prevent overlap and remove spines
plt.tight_layout()
sns.despine()

# Display the plot
plt.show()
```

```python
# Import necessary libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create a sample DataFrame with an additional column
data = {
    'Category': ['numbers']*20 + ['additional']*20,
    'Value': [1, 2, 3, 4, 5, 1, 2, 2, 3, 4, 5, 5, 5, 3, 2, 1, 4, 5, 3, 2] + \
             [1, 3, 2, 5, 4, 2, 3, 1, 5, 4, 3, 1, 4, 2, 5, 3, 2, 1, 4, 5]
}
df_long = pd.DataFrame(data)

# Create the count plot with shared x-axis
sns.countplot(x='Value', hue='Category', data=df_long, palette='viridis')

# Enhancing the plot
plt.title('Count of Values in Columns')
plt.ylabel('Count')
plt.xlabel('Value')

# Adding a vertical line to visually separate the categories
max_count = df_long['Value'].value_counts().max()
plt.axvline(x=4.5, color='grey', linestyle='--', ymax=max_count)

# Display the plot
plt.show()
```
